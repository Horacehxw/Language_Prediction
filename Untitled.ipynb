* use the proportion change in Canada for English and French users to get the required parameters a, c

where a captures the trasition probability,  c captures the language growth speed.

* We define the l2 loss on the statistic's year and use PSO to optimize the parameter.

```python
pop_canada_after_1971 = pop_canada[(pop_canada['Time']>=1971)]['PopTotal'].values
```

```python
# this data only have 5 year range statistics
ratio_canada_45year = np.array([[64.6, 20.1, 12.5],
                                [64.13, 19.3,13.57],
                                [65.4, 17.9, 15.35],
                                [65.7, 17.1, 16.25],
                                [65.6, 16.2, 17.14],
                                [66.1, 15.4, 17.5],
                                [66.8, 14.7, 18.04],
                                [67.9, 13.5, 17.4],
                                [68.1, 12.6, 17.5],
                                [68.3, 11.9, 17.9]
                               ])
# normalize the proportion along each year
ratio_canada_45year = np.apply_along_axis(func1d=lambda x: x/np.sum(x), axis=1, arr=ratio_canada_45year)
```

```python
from numpy import linalg as LA
def cost(x):
    '''
    return the l1 loss of given parameter a,c
    '''
    c, a = x[0], x[1]
    canada = Country("English", "French", 63.6,20.1,12.5,np.exp(0.778),np.exp(0.318),
                     0.5,pop_canada_after_1971, c=c, a=a)
    loss = 0
    for i in range(50):
        if i % 5 ==0:
            #print("time={:5}:\tx={:1.5f}, y={:1.5f}, z={:1.5f}".format(canada.time, canada.x, canada.y, canada.b))
            j = int(i/5)
            fake = np.array([canada.x, canada.y, canada.b])
            real = ratio_canada_45year[j,:]
            loss += LA.norm(fake-real ,2)
        canada.update()
        canada.grow()
    return loss
```

```python
from pyswarm import pso
lower_bound = [0, 1] #c, a
upper_bound = [1, 10] #c, a
xopt, fopt = pso(cost, lower_bound, upper_bound)
```

```python
class World():
    '''
    Contains list of Country Object and a migration graph between countries.
    
    Impelment the update, transmit and grow for each country
    '''
    def __init__(self,pop, G=G_migration, c=xopt[0], a=xopt[1]):
        '''
        build the list of Country objects --> countries

        the order of elements follows the "list_country" object
        '''
        self.time = 0
        self.G = G
        self.countries = [] 
        for country_name in list_country:
            # create the list of population of every year, start from 1961 inclusive
            pop_this = list(pop[pop['Location']==country_name].PopTotal.values)
            x, y, name_X, name_Y = country_dict[country_name]
            idx, idy = list_language.index(name_X), list_language.index(name_Y)
            k = similarity[idx, idy]
            Ix, Iy = language.iloc[idx].centrality, language.iloc[idy].centrality
            self.countries.append(Country(name_X, name_Y, x, y, 0, Ix, Iy, k, pop_this, a=a, c=c))
            
    def update(self):
        '''
        update each edge of migration
        update --> transmit --> grow
        '''
        for country in self.countries:
            country.update()
        for edge in self.G.edges(data=True):
            id1, id2, weight = edge[0], edge[1], edge[2]['weight']
            country1, country2 = self.countries[id1], self.countries[id2]
            country1.transmit(country2, weight)
        for country in self.countries:
            country.grow()
        self.time += 1
        
    def report_language(self, languages, native=False):
        pop_language = []
        for language in languages:
            pop_ = 0
            for country in self.countries:
                if country.name_X == language:
                    if native:
                        pop_ += country.X + country.B * country.x / (country.x+country.y)
                    else:
                        pop_ += country.X + country.B
                elif country.name_Y == language:
                    if native:
                        pop_ += country.Y + country.B * country.y / (country.x+country.y)
                    else:
                        pop_ += country.Y + country.B
            pop_language.append(pop_)
        return pop_language
```

## Step 5: Power Language Index --> Power Country Index

```python
target_language_2025= ['Hindustani', 'Bengali','Spanish', 'Javanese', 'Urdu', 'Portuguese']
target_PLI_2025 = [0.122, 0.029, 0.329, 0.04, 0.119, 0.026]
```

```python
def power_country(target_language, target_PLI, year):
    world = World(pop[pop['Time']>=2010])
    for year in range(2010, year):
        world.update()
        language_data_native.append(world.report_language(list(language.dropna(how='any').Language.values), native=True))
    country_score = {}
    for j in range(len(list_country)):
        score = 0
        country = world.countries[j]
#         if (list_country[j] == "Congo"):
#             import pdb; pdb.set_trace()
        if country.name_X in target_language or country.name_Y in target_language:
            pass
        if country.name_X in target_language:
            score += (country.x + country.b) * np.exp(target_PLI[target_language.index(country.name_X)]) * country.P
        if country.name_Y in target_language:
            score += (country.y + country.b) * np.exp(target_PLI[target_language.index(country.name_Y)]) * country.P
        if score != 0:
            country_score[list_country[j]] = score
    return country_score
```
